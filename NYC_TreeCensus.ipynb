## Importing relevant libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Creating functions 

### 1. getData() Function when called, returns a dataframe with stripped white spaces

```python
def getData():
    
    #Pulling the data
    path = 'C:/Users/makye/github/ANT-Theano-2-27/additional_resources/datasets/NYC Trees/data_100000.csv'
    df = pd.read_csv(path)
    
    #Stripping whitespace from cells that have a stringlike object in them
    df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
    
    return df

df = getData()

df
```

### 2. exploreData() Function when called, returns exploratory information on the dataFrame.

```python
def exploreData(df):

    #checking structure of dataset, variable dtypes, count of non-null values per variable
    print(df.info())

    #list of variables/columns
    print(df.columns)
    
    #checking/counting missing values per variable
    print(df.isnull().sum(axis = 0))  
    

exploreData(df)
```

### 3. getMissing() Function when called, returns a dataframe with missing data percentage by feature and plots it

```python
""" getMissing() Function when called, returns a dataframe with missing data percentage by feature and plots it"""

def getMissing(df):
    
    #Generating percentages of missing data
    missing_var = 100*df.isnull().sum()/df.shape[0]
    missing_data = missing_var.sort_values(ascending = True)
    
    #Plotting percentages of missing data
    plot = missing_data.plot(kind='barh',figsize=(20,9.5), grid = False,
                             label="Percentage",color='lightgreen',fontsize=12)
    plot.set_xlabel("% of Missing Values in Data")
    plt.figtext(.5,0.9,'Missingness of Data By Variable',fontsize=30,ha='center')

    return missing_data

getMissing(df)
```

### 4. cleanData() Function when called, returns a fairly cleaned out dataFrame for further data analysis


```python
def cleanData(df):
    
    #Replacing nan values with dead for attribute 'health'
    df1 = df.replace(np.nan, 'Dead', regex=True) 
  

    #check for duplicates
    df1.duplicated().any()
    
    return df1

df1 = cleanData(df)
df1
```

### 5. createCsv() Function when called Saves cleaned out dataframe to csv 

```python
def createCsv(df1):
    
    path = 'C:/Users/makye/github/NYC_TreeCensus/data_clean_GOOD_ENOUGH.csv'
    df1.to_csv(path, index=False)
    
createCsv(df1)
```
